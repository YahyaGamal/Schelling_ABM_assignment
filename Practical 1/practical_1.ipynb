## Predator-prey model desciption <a id='1'></a>

The predator-prey model is an abstract model that explores the dynamics between predators, preys and prey food availability on the populations of each species.
The predator-prey Agent Based Moodel (ABM) includes: a space, and agents and objects that temporally interact with each other and with the model space.
We describe each component of the predator-prey ABM hereafter.

### Space
The space is represented as a grid of square cells. Each grid cell represents a patch of land that can include grass (food for prey). The grid cells have a `grass-growth-rate` representing the time needed to regrow a consumed grass in a land patch.

### Agents
The model includes two types of agents: predators and preys.

Predators (wolves) and preys (sheep) have the following state parameters (i.e., variables that dictate the state of the agent during the simulation):
1. `energy`            ; the energy acquired from food (preys eat grass and predators eat preys)
2. `reproduction-rate` ; the rate at which the agent generates a child of the same type

Agents aim to keep their `energy > 0` to avoid dying. The agents move randomly in space (1 land patch step at a time) to search for food. After each move, the agents lose 1 unit of energy.

Sheep consume grass whenever they move to a patch with grown grass, and wolves consume preys whenever they move to a patch with a sheep in it.

After consuming food, sheep and wolves increase their energy by an input rate. They reproduce at an input rate given that their `energy > 0` -- they generate a similar agent at the same location. They die if their `energy < 0`.


### Temporality
Each time step in the model represents an iteration where each agent moves randomly (1 patch distance), checks for food and reproduces.

## Explore the predator-prey model <a id='2'></a>

To explore the predator-prey model, we provide an ABM version that runs in a Python notebook. 
The models runs by calling the `run_predator_prey_model()` function, which has the following input parameters:
1. `steps`: the number of time steps the model runs
2. `initial_sheep`: the total number of sheep randomly distributed in space
3. `initial_wolves`: the total number of wolves randomly distributed in space
4. `sheep_reproduce`: the reproduction rate of sheep
5. `wolf_reproduce`: the reproduction rate of wolves
6. `sheep_gain_from_food`: the energy gained from consuming one unit of grass
7. `wolf_gain_from_food`: the energy gained from consuming one prey
8. `grass_regrowth_time`: the time steps taken to regrow grass

It should be noted that:
1. The type of the agent is represented visually by the colours 
(<span style="background-color: white">white for sheep</span> and <span style="background-color: #808080">black for wolves</span>)
2. The availability of grass in vacant patches is represented by the colours (<span style="background-color: green">green for available grass</span> and <span style="background-color: #8B4513">brown for no grass</span>)
2. The ratio between the two different types of agents is always 0.5 (50% of the agents are yellow and the rest are blue)
3. At inialisation, the exact locations of agents are randomised while maintaining the required input `density`. This implies that different spatial patterns may be observed across different runs using the same input parameters.
